```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm
import seaborn as sns
import datetime
```

```python
lp_n=pd.read_csv('neverfullmm.csv')
comp_n=pd.read_csv('price_scrape_neverfull.csv')
```

```python
pd.to_numeric(comp_n['Price'])
```

```python
fp_n=comp_n[comp_n['Seller']=='Fashionphile']
rb_n=comp_n[comp_n['Seller']=='Rebag']
rr_n=comp_n[comp_n['Seller']=='TRR']
```

```python
rb_n['Condition'].value_counts()
```

```python
fp_n['Condition'].value_counts()
```

```python
mono_fp=(fp_n[(fp_n['Description'].str.contains('Monogram'))
      &(fp_n['Condition'].isin(['Condition: Very Good', 'Condition: Good']))])
mono_fp
```

```python
mono_rb=(rb_n[(rb_n['Description'].str.contains('Monogram Canvas MM'))
              &(~rb_n['Description'].str.contains('Limited'))
      &(rb_n['Condition'].isin(['Good Condition', 'Very Good Condition']))])
```

```python
print('the avg price for a neverfull mm bag on rebag: ${:.2f}'.format(mono_rb['Price'].mean()))
print('the avg price for a neverfull mm bag on fashionphile: ${:.2f}'.format(mono_fp['Price'].mean()))
print('the avg price for a neverfull mm bag on leprix: ${:.2f}'.format(lp_n['Price'].mean()))
```

```python
lp_s=pd.read_csv('speedy30.csv')
comp_s=pd.read_csv('price_scrape_speedy.csv')
```

```python
fp_s=comp_s[comp_s['Seller']=='Fashionphile']
rb_s=comp_s[comp_s['Seller']=='Rebag']
```

```python
s_mono_fp=(fp_s[(fp_s['Description']==('Monogram Speedy 30'))
      &(fp_s['Condition'].isin(['Condition: Very Good', 'Condition: Good']))])
s_mono_fp.head()
```

```python
s_mono_rb=(rb_s[(rb_s['Description'].str.contains('Monogram'))
      &(rb_s['Condition'].isin(['Good Condition', 'Very Good Condition']))&
               (~rb_s['Description'].str.contains('Multicolor'))])
s_mono_rb.head()
```

```python
print('the avg price for a speedy 30 bag on rebag: ${:.2f}'.format(s_mono_rb['Price'][218]))
print('the avg price for a speedy 30 bag on fashionphile: ${:.2f}'.format(s_mono_fp['Price'].mean()))
print('the avg price for a speedy 30 bag on leprix: ${:.2f}'.format(lp_s['Price'].mean()))
```
