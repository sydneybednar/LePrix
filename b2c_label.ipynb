# b2c partner labeling

```python
import pandas as pd
import numpy as np
```

```python
client=pd.read_excel('source_clients.xlsx')
part=pd.read_excel('b2c_partners.xlsx')
```

```python
part['username']=part['username'].str.lower()
```

```python
buy_list=list(part['username'])
```

```python
client.loc[client['username'].isin(buy_list), 'medium']='B2C Store Partner'
```

```python
client[client['medium']=='B2C Store Partner']
```

```python
# client.to_excel('b2c_sources.xlsx')
```

```python
gmv=pd.read_excel('client_gmv.xlsx')
gmv['Retailer Name']=gmv['Retailer Name'].str.lower()
gmv.head()
```

```python
client=client.merge(gmv, left_on='username', right_on='Retailer Name', how='left')
client.head()
```

```python
client.to_excel('b2c_sources.xlsx')
```
